```python
import numpy as np
from random import randrange

def generate_mock_data(number_of_cases, start_date, end_date, drugs, origins=[], sellers=[]):
    listings = []
    for i in range(number_of_cases):
        date = random_date(start_date, end_date)
        drug = drugs[np.random.randint(0, len(drugs)-1)]
        if origins:
            origin = origins[np.random.randint(0, len(origins)-1)]
        if sellers:
            seller = sellers[np.random.randint(0, len(sellers)-1)]
        listing = {
            "drug": drug,
            "origin": origin,
            "date": date,
            "origin": origin,
            "seller": seller
        }
        listings.append(listing)
    return listings
    
def random_date(start, end):
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

```python
from datetime import datetime, timedelta
drugs = [i for i in range(1,11)]
origins = [i for i in range(1,11)]
sellers = [i for i in range(1,11)]
start_date = datetime.strptime('1/1/2018 1:30 PM', '%m/%d/%Y %I:%M %p')
end_date = datetime.strptime('1/1/2019 4:50 AM', '%m/%d/%Y %I:%M %p')

listings = generate_mock_data(100, start_date, end_date, drugs, origins, sellers)
```
